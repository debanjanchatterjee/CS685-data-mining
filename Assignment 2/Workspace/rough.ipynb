```python
import csv
import networkx as nx
```

```python
read=list()

f = open("shortest-path-distance-matrix.txt", "r")
for x in f:
  read.append(x.strip())

read=read[17:]
```

```python
print(read[0])
```

```python
G = nx.Graph()

edge_set=set()

p=1
for i in read:
    q=1
    
    for j in i:
        if j=='1':
            e=[p,q]
            e.sort()
            edge_set.add(tuple(e))
        else:
            q=q+1
            continue
        q=q+1
    p=p+1


```

```python
print(len(edge_set))
```

```python
edge_list=list()



for x in edge_set:
    edge=list(x)
    edge_list.append(edge)
    
print(len(edge_list))
```

```python
G.add_edges_from(edge_list)
```

```python
numbercc=nx.algorithms.components.number_connected_components(G)
print(numbercc)
```
