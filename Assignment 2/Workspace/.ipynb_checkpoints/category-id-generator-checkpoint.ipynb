```python

def convertLS(l):
    s=""
    for i in l:
       s=s+i+'.' 
    return s[:-1]


m_new=dict()



for keys in m:
    l=m[keys]
    if keys not in m_new:
        m_new[keys]=list()
    for x in l:
        new_s=convertLS(x)
        if new_s not in m_new[keys]:
            m_new[keys].append(new_s)
    

print(len(m_new))
print(m_new)
    
```
