```python
import csv
import networkx as nx
```

```python
with open('paths_finished.tsv', newline='') as f:
    reader = csv.reader(f)
    pfread = list(reader)
f.close()

pfread=pfread[16:]
```

```python
with open('article-ids.csv', newline='') as f:
    reader = csv.reader(f)
    article = list(reader)
f.close()
```

```python
with open('category-ids.csv', newline='') as f:
    reader = csv.reader(f)
    category = list(reader)
f.close()
```

```python
with open('article-categories.csv', newline='') as f:
    reader = csv.reader(f)
    ac = list(reader)
f.close()
```

```python
a_map=dict()
for i in article:
    a_map[i[0]]=i[1]
    
print(a_map)
```

```python
c_map=dict()

for i in ac:
    c_map[i[0]]=i[1:]

print(c_map)
```

```python

cat_map=dict()
cat_map_rev=dict()
for i in category:
    cat_map[i[1]]=i[0]
    cat_map_rev[i[0]]=i[1]
print(cat_map)
print(cat_map_rev)
```

```python
with open('edges.csv', newline='') as f:
    reader = csv.reader(f)
    edge_list= list(reader)
f.close()
```

```python
s=set()
isolated_nodes=set()
for i in edge_list:
    for j in i:
        s.add(j)

for c in range(1,4605):
    id=str(c)
    id=id.zfill(4)
    id='A'+id
    if id not in s:
        isolated_nodes.add(id)
print(isolated_nodes)
```

```python
G = nx.DiGraph()
G.add_edges_from(edge_list)
```

```python
def subCategories(c):
    c_name=cat_map[c]
    c_arr=c_name.split('.')
    str=''
    r_list=list()
    for s in c_arr:
        str=str+s
        str_id=cat_map_rev[str]
        r_list.append(str_id)
        str=str+'.'
    return r_list
        

def findSC(l):
    lnew=list()
    for i in l:
        lnew=lnew+subCategories(i)
    return lnew
    
```

```python
def modify(l):
    if '<' not in l:
        return l
    index=-1
    for i in range(0,len(l)):
        if l[i]=='<':
            index=i
            break
    
    temp1=l[:index-1]
    temp2=l[index+1:]
    l=temp1+temp2
    return modify(l)


result=list()


for line in pfread:
    line_arr=line[0].split('\t')
    s=line_arr[-2]
    #if '<' in s:
    #    continue
    
    #s=s.strip()
    #x=s.find('\t')
    #s=s[:x]
    s_arr=s.split(';')
    #print(s_arr)
    cat_list=list()
    cat_set=set()
    s_arr=modify(s_arr)
    if len(s_arr)<=1:
        continue
    
    source=s_arr[0]
    destination=s_arr[-1]
    
    #if source not in a_map:
    #    source=artc[source]
        
    #if destination not in a_map:
    #    destination=artc[destination]
        
    aid_src=a_map[source]
    aid_dst=a_map[destination]
    
    
    cl_src=c_map[aid_src]
    cl_dst=c_map[aid_dst]
    
    
    cat_list_src=findSC(cl_src)
    cat_list_dst=findSC(cl_dst)
    
    
    lhp=len(s_arr)-1
    
    sa=list()
    for a in s_arr:
        aid=a_map[a]
        sa.append(aid)
    #print(sa)   
    if sa[0] in isolated_nodes:
        continue
    if sa[-1] in isolated_nodes:
        continue
    shortest_path=nx.shortest_path(G, source=sa[0], target=sa[-1])
    
    lsp=len(shortest_path)-1
    ratio=lhp/lsp
    
    for x in cat_list_src:
        for y in cat_list_dst:
            e=[x,y,ratio]
    result.append(e)
    
    
result=sorted(result, key=lambda x:(x[0],x[1]))
print(result)
```

```python
file = open('category-ratios.csv', 'w+', newline ='') 
with file:     
    write = csv.writer(file) 
    write.writerows(result) 
file.close()
```
