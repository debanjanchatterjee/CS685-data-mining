```python
articleid = 'article-ids.csv'
article_ids = pd.read_csv(articleid,names=["articleId","article"])
```

```python
articleidlist=article_ids['articleId'].values.tolist()
```

```python
len(articleidlist)
```

```python
artid_dict=article_ids.set_index('article').T.to_dict('records')

artid_dict=artid_dict[0]
```

```python
artid_dict
```

```python
artspdict=dict()#article shortest path dict
```

```python
fil="wikispeedia_paths-and-graph/wikispeedia_paths-and-graph/shortest-path-distance-matrix.txt"
with open(fil) as f:
        j=0
        for line in f:
            artspdict[articleidlist[j]]=line
            j+=1
```

```python
zl=lol=0
with open('finished-paths-back.csv', mode='a',newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["path","humanlength","shortestpath","hl/spratio"])
    for i in finpathlist:
        path=i.split(";")
        #humanlength=len(path)
        #if "<" in path:
            #continue
        #else:
        humanlength=len(path)-1
        pathstart=path[0]
        pathend=path[humanlength]
        articlestartid=artid_dict[pathstart]
        articleendid=artid_dict[pathend]
        articleendid=int(articleendid[1:])
        shortestpath=artspdict[articlestartid][articleendid-1]
        if shortestpath == '_' :
            ratio='nan'
            shortestpath=0
            #lol+=1
           # print(path)
        elif shortestpath == '0':
                ratio='0'
                shortestpath=0
            #    zl+=1
           #     print(path,shortestpath)
            #else:
            #print(pathstart,"-",pathend,"",shortestpath)
        else:
            #print(path,humanlength,shortestpath)
            ratio=round(humanlength/int(shortestpath),2)
        csv_writer.writerow([i,humanlength,shortestpath,ratio])
            
        
    
```

```python
with open('finished-paths-no-back.csv', mode='a',newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["path","humanlength","shortestpath","hl/spratio"])
    for i in finpathlist:
        path=i.split(";")
        
        stack = []
        for j in path:
            if j !='<':
                stack.append(i)
            elif j=='<':
                stack.pop()
        
        
        humanlength=(len(stack)-1)
        pathstart=path[0]
        pathend=path[len(path)-1]
        articlestartid=artid_dict[pathstart]
        articleendid=artid_dict[pathend]
        articleendid=int(articleendid[1:])
        shortestpath=artspdict[articlestartid][articleendid-1]
        if shortestpath == '_' :
            ratio='nan'
            shortestpath=0
            #lol+=1
           # print(path)
        elif shortestpath == '0':
                ratio='0'
                shortestpath=0
            #    zl+=1
           #     print(path,shortestpath)
            #else:
            #print(pathstart,"-",pathend,"",shortestpath)
        else:
            #print(path,humanlength,shortestpath)
            ratio=round(humanlength/int(shortestpath),2)
        
        
        csv_writer.writerow([i,humanlength,shortestpath,ratio])
    
```
